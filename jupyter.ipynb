### Normal distribution

The normal distribution has the following functional form:
$$ f(x,\mu,\sigma) = \frac{1}{\sigma \sqrt{2\pi}} \exp \left( -\frac{(x-\mu)^2}{2\sigma^2 } \right) $$

```python
fig,(ax1,ax2,ax3) = plt.subplots(3,1)
fig.suptitle(r'Normal distribution with $\mu=0$ and $\sigma=1$')

ax1.plot(np.random.normal(0,1,1000),'.',alpha=0.3)
ax1.set_ylabel('$x_i$')
ax1.set(xticklabels=[])  

x=np.linspace(-4,4,201)
ax2.plot(x, norm.pdf(x,0,1) , linewidth=2, color='b',label='pdf')
ax2.set_ylabel('pdf')
ax2.grid()

ax3.plot(x, norm.cdf(x,0,1), linewidth=2, color='r',linestyle='-',label='cfd')
ax3.set_xlabel('$x$')
ax3.set_ylabel('cdf')
ax3.set_ylim(0,1)
ax3.grid()

plt.show()
```
