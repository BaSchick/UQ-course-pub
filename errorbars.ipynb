```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
x = np.arange(10)
y = 2.5 * np.sin(x / 20 * np.pi)
yerr = np.linspace(0.05, 0.2, 10)

plt.errorbar(x, y + 3,yerr=yerr,capsize=4, label='data with error bars')

plt.legend()
plt.show()
```

Be sure to always say what the error bars mean as there is no standard definition. Is it a 95%-CI, the standard error or the standard deviation? Between the 95%-CI and the standard error, there is a nearly factor 2 (1.96 for normal distribution) difference!

```python
fig,ax = plt.subplots(1,1)
x = np.arange(10)
y = 2.5 * np.sin(x / 20 * np.pi)
yerr = np.linspace(0.05, 0.2, 10)
plt.plot(x,y,label='data')
ax.fill_between(x,y-yerr,y+yerr,facecolor='blue',alpha=0.1,label='CI')
plt.legend()
plt.show()
```
